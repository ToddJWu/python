```python
import requests as req
import os, json
from bs4 import BeautifulSoup as bs

# 隨機取得 User-Agent
from fake_useragent import UserAgent
ua = UserAgent() # cache=True 表示從已經儲存的列表中提取
```

```python
# ua.random
```

```python
# 建立貼圖資料夾
filePath = "LineStickers"
if not os.path.exists(filePath):
    os.mkdir(filePath)
    print("Create File for images.")
else:
    print("File Exists.")
```

```python
list_stickers = []
```

```python
headers = {
    "user-agent": ua.random
}
```

```python
url = "https://store.line.me/stickershop/product/26924/zh-Hant"
res = req.get(url, headers=headers)
soup = bs(res.text, "lxml")
```

```python
soup
```

```python
elements = soup.select("ul.mdCMN09Ul.FnStickerList > li.mdCMN09Li.FnStickerPreviewItem")

for e in elements:
#     print(e["data-preview"])
    obj = json.loads(e["data-preview"])
    
    list_stickers.append(obj)
```

```python
print(list_stickers)
```

```python
for sticker in list_stickers:
    # 用 curl 下載
    os.system(f"curl {sticker['staticUrl']} -o {filePath}/{sticker['id']}.png")
    print(f"{sticker['id']} Downloaded!")
```
