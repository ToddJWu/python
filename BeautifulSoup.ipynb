## Cookies

```python
# PTT sex (西斯) 板
url = "https://www.ptt.cc/bbs/sex/index.html"

# 首頁網址
prefix = 'https://www.ptt.cc'

# 設定 cookie
my_cookies = {
    "over18": "1"
}

# 用 requests 的 get 方法把網頁抓下來
res = requests.get(url, cookies = my_cookies) 

# 指定 lxml 作為解析器
soup = bs(res.text, "lxml")
```

```python
for a in soup.select("div.r-ent > div.title > a"):
#     print(a)
    print(a.get_text())
    print(prefix + a.get("href"))
```

## Example

```python
# 取得新聞列表
url = "https://www.iiiedu.org.tw/category/news/" 

# 用 requests 的 get 方法把網頁抓下來
res = requests.get(url) 

# 指定 lxml 作為解析器
soup = bs(res.text, "lxml") 
```

```python
# 取得 標題 跟 連結
news_title = []
news_url = []

for a in soup.select("a[rel='bookmark']"):
    print(a.get_text())
    news_title.append(a.get_text())
    news_url.append(a["href"])
```

```python
len(news_url),len(news_title)
```

```python
for title, url in zip(news_title, news_url):
    res = requests.get(url)
    context = bs(res.text, "lxml")
    
    print(title)
    print("-"*73)
    print(context.select_one("div.entry-content").get_text())
    print("-"*73)
```
